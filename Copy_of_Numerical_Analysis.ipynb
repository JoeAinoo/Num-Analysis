Load the Iris Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names
```

2. Separate the Data for Iris Virginica

```python
# Filter data for Iris Virginica
virginica_data = data[target == 2]
petal_length_virginica = virginica_data[:, 2]  # Petal Length
petal_width_virginica = virginica_data[:, 3]   # Petal Width
```

3. Calculate the Regression Coefficients

```python
# Calculate the means of petal length and petal width
mean_petal_length = np.mean(petal_length_virginica)
mean_petal_width = np.mean(petal_width_virginica)

# Calculate the slope (m) and intercept (b)
numerator = np.sum((petal_length_virginica - mean_petal_length) * (petal_width_virginica - mean_petal_width))
denominator = np.sum((petal_length_virginica - mean_petal_length)**2)
slope = numerator / denominator
intercept = mean_petal_width - slope * mean_petal_length

# Equation of the least square regression line
print(f"Equation of the regression line: y = {slope:.4f}x + {intercept:.4f}")
```

4. Plot the Data and Regression Line

```python
# Plot the data points
plt.scatter(petal_length_virginica, petal_width_virginica, color='blue', label='Iris Virginica')

# Plot the regression line
regression_line = slope * petal_length_virginica + intercept
plt.plot(petal_length_virginica, regression_line, color='green', label='Regression Line')

# Labels and legend
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width for Iris Virginica')
plt.legend()
plt.show()
```

```python
# Predict petal width for a petal length of 4.2 cm
predicted_petal_width = slope * 4.2 + intercept
print(f"Predicted petal width for petal length 4.2 cm: {predicted_petal_width:.4f} cm")
```
